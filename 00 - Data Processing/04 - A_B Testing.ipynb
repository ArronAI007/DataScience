    * Test whether sample differences can be generalized as population differences
    * The null hypothesis states that the mean of all groups is equal
    * The alternative hypothesis states that at least one of the means is different
    * Use F statistic to compute the P-Value using the F distribution with df1 and df2 degrees of freedom
    * If the p-value is below Î¼, we reject the null hypothesis, this means that at least one group mean is significantly different
